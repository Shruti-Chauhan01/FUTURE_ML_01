```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset
df = pd.read_csv('Superstore_Sales.csv', encoding='ISO-8859-1')

# Step 3: Quick look at the data
print(df.shape)
print(df.columns)
df.head()
```

```python
# Check missing values
df.isnull().sum()

# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract useful time features
df['Month'] = df['Order Date'].dt.to_period('M')  # For monthly sales
df['Year'] = df['Order Date'].dt.year
df['DayOfWeek'] = df['Order Date'].dt.dayofweek
```

```python
# Group sales by Month
monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()
monthly_sales['Month'] = monthly_sales['Month'].astype(str)
monthly_sales.columns = ['ds', 'y']  # for Prophet model later
monthly_sales.head()
```

```python
# Step 1: Import Prophet
from prophet import Prophet

# Step 2: Make sure 'ds' and 'y' columns are correct
monthly_sales['ds'] = pd.to_datetime(monthly_sales['ds'])  # ensure datetime type

# Step 3: Initialize the Prophet model
model = Prophet()

# Step 4: Fit the model to your data
model.fit(monthly_sales)

# Step 5: Create future dataframe (e.g., predict next 6 months)
future = model.make_future_dataframe(periods=6, freq='M')

# Step 6: Forecast future sales
forecast = model.predict(future)

# Step 7: View predicted results
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(6)
```

```python
model.plot(forecast);
```

```python
model.plot_components(forecast);
```

```python
forecast[['ds', 'yhat']].rename(columns={'yhat': 'Predicted_Sales'}).to_csv('forecast_output.csv', index=False)
```
